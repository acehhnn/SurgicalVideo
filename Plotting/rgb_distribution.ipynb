```python
sample_real_blood = real_blood.sample(n=n_min)
sample_real_nonblood = real_nonblood.sample(n=n_min)
sample_fake_blood = fake_blood.sample(n=n_min)
sample_fake_nonblood = fake_nonblood.sample(n=n_min)

rgb_real_blood = sample_real_blood[['R','G','B']]
rgb_real_nonblood = sample_real_nonblood[['R','G','B']]
rgb_fake_blood = sample_fake_blood[['R','G','B']]
rgb_fake_nonblood = sample_fake_nonblood[['R','G','B']]
```
